```python
#find percentage of null values of each column in given data frame df(train.csv)
#formula  : df.isnull().sum()/no. of records *100
#How to find no. of records/rows :  
df.shape
```

```python
#find the no. of records means rows
r=df.shape[0]
print(r)
```

```python
#find the no. of columns 
c=df.shape[1]
print(c)
```

```python
null_var=df.isnull().sum()/df.shape[0] *100
print(null_var) #value show in %
```

```python
d=null_var[null_var>20].keys()   #keys() inbuilt function access column
#to access those column whose null values in % >20 and hold in d
print(d)
```

```python
#to delete column which hold in d object 
#to delete a column or  row then use drop() inbuilt function
df_new=df.drop(columns=d) #or df.drop(d,axis=1)
```

```python
df_new.shape #to check shape of new dataframe after delete the column
```

```python
#in LotFrontage column , 17.7 % null values which less than 20%
df_new["LotFrontage"].unique()
```

```python
df_new.info()  #check datatype of LotFrontage column
```

```python
#mean(),mode() and median() always apply on numerical column 
df_new["LotFrontage"].median()
```

```python
df_new["LotFrontage"].mean()
```

```python
df_new["LotFrontage"].mode()
```

```python
#Fill Null values of LotFrontage column with median of LotFrontage column 
#permanently 
df_new["LotFrontage"].fillna(df_new["LotFrontage"].median(),inplace=True)
#inplace=True  means permanenly change in dataframe df_new 
#fillna()  : only fill null values 
```

```python
#check
df_new.isnull().sum()
```

```python
#visualise 
plt.figure(figsize=(25,20)) #width=25 and height=20
sns.heatmap(df_new.isnull())
plt.show()
```
