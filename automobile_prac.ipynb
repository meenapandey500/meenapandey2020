# Regularization

```python
from sklearn.linear_model import Lasso , Ridge
for i in range(200,501,50):
    l1=Lasso(i)
    l1.fit(X_train,Y_train)
    print("Lambda=",i)
    print(l1.score(X_train,Y_train))
    print(l1.score(X_test,Y_test))
    print()
    
```

```python
for i in range(0,51,1):
    l2=Ridge(i)
    l2.fit(X_train,Y_train)
    print("Lambda=",i)
    print(l2.score(X_train,Y_train))
    print(l2.score(X_test,Y_test))
    print()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
L1=Lasso(200)

cross_val_score(L1,X,Y,cv=4)
```

```python
L2=Ridge(2)
cross_val_score(L2,X,Y,cv=4).mean()
```
