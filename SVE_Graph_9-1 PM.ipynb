```python
#SVM example 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
```

```python
# linear data
X = np.array([1, 5, 1.5, 8, 1, 9, 7, 8.7, 2.3, 5.5, 7.7, 6.1]) #input
y = np.array([2, 8, 1.8, 8, 0.6, 11, 10, 9.4, 4, 3, 8.8, 7.5])#input
#Here X and Y both are independent variable means features means columns
```

```python
# show unclassified data
plt.figure(figsize=(7,7))
plt.scatter(X, y)
plt.xticks(X)
plt.yticks(y)
plt.show()
```

```python
# shaping data for training the model
training_X = np.vstack((X, y)).T    #vertical stack vstack() inbuilt function of numpy
#stack is a temporary memory of system 
training_y = [0, 1, 0, 1, 0, 1, 1, 1, 0, 0, 1, 1] #output
```

```python
training_X
```

```python
#Now we can create the SVM model using a linear kernel. SVC inbuilt class which define svm
# define the model
clf = svm.SVC(kernel='linear', C=1.0) #C =1.0 and clf is a user defined object
```

```python
# train the model
clf.fit(training_X, training_y)
```

```python
# get the weight values for the linear equation from the trained SVM model
w = clf.coef_[0]
# get the y-offset for the linear equation
a = -w[0] / w[1]

# make the x-axis space for the data points
XX = np.linspace(0, 13)

# get the y-values to plot the decision boundary
yy = a * XX - clf.intercept_[0] / w[1]

# plot the decision boundary
plt.plot(XX, yy, 'k-') #k means algo

# show the plot visually
plt.scatter(training_X[:, 0], training_X[:, 1], c=training_y)#c means color
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import svm
```

```python
# non-linear data
circle_X, circle_y = datasets.make_circles(n_samples=300, noise=0.05)
```

```python
# show raw non-linear data
plt.scatter(circle_X[:, 0], circle_X[:, 1], c=circle_y, marker='.')
plt.show()
```

```python
# make non-linear algorithm for model
nonlinear_clf = svm.SVC(kernel='rbf', C=1.0) #radial basis and C ==>error provides 
```

```python
 #training non-linear model
nonlinear_clf.fit(circle_X, circle_y)
```

```python
# Plot the decision boundary for a non-linear SVM problem
def plot_decision_boundary(model, ax=None):
    if ax is None:
        ax = plt.gca() #gca() get current axis  which is defiend in matplotlib
        
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x) #meshgrid() inbuilt method of numpy to draw a rectangle grid 
    #grid means horizontal and vertical line

# shape data
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    
# get the decision boundary based on the model
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary
    ax.contour(X, Y, P,
               levels=[0], alpha=0.5,
               linestyles=['-'])
```

```python
# plot data and decision boundary
plt.scatter(circle_X[:, 0], circle_X[:, 1], c=circle_y, s=50)#s means size of datapoint
#call  function
plot_decision_boundary(nonlinear_clf)
plt.scatter(nonlinear_clf.support_vectors_[:, 0], nonlinear_clf.support_vectors_[:, 1], s=50, 
            lw=1, facecolors='none')
plt.show()
```
