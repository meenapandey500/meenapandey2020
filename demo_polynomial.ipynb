```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Demo of polynomial regression 
#we create the input variable X with help of arange() of numpy
X=np.arange(-100,100) #start=-100  stop=100-1=99 and step=+1
print(X)
```

```python
#create function for linear :- equation of line Y=mX+b here m=slope or coef
#b intercept
#linear means stright line means degree of X=1 means power of X=1
def poly_1(X) : #poly_1() user defined function which degree=1 of X
    #this function is passing argument and return value
    m=4  #assign slope
    b=5  #assign intercept
    Y=m*X+b
    return Y
```

```python
#Quadradic means degree of X=2 means create a curve
#equation of Quadratic Y=m1X^2+m2*X^1+b
def poly_2(X):#poly_2() user defined function which degree=2 of X
    #this function is passing argument and return value
    m1=6
    m2=4
    b=5
    Y=m1*(X**2)+m2*X+b
    return Y
```

```python
#Cubic means degree of X=3 means create a curve
#equation of cubic Y=m1X^3+m2*X^2++m3*X^1+b
def poly_3(X):#poly_3() user defined function which degree=3 of X
    #this function is passing argument and return value
    m1=8
    m2=6
    m3=4
    b=5
    Y=m1*(X**3)+m2*(X**2)+m2*X+b
    return Y
```

```python
#call in main program
Y=poly_1(X)
print(Y)
#visualise a graph between X and Y here X=input and Y=input
plt.plot(X,Y)
plt.xlabel("Value of X")
plt.ylabel("Value of Y")
plt.show()
```

```python
#call in main program
Y=poly_2(X)
print(Y)
#visualise a graph between X and Y here X=input and Y=input
plt.plot(X,Y)
plt.xlabel("Value of X")
plt.ylabel("Value of Y")
plt.show()
```

```python
#call in main program
Y=poly_3(X)
print(Y)
#visualise a graph between X and Y here X=input and Y=input
plt.plot(X,Y)
plt.xlabel("Value of X")
plt.ylabel("Value of Y")
plt.show()
```
