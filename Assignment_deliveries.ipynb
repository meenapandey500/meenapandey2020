```python
#1) Total runs of Virat Kohli (including all matches)
total=df.groupby("batsman")["total_runs"].sum()
total_kohali=total["V Kohli"]
print("Total runs of Virat Kohli=  ",total_kohali)
```

```python
#2) Total wickets and their dismissal types by YS Chahal
wickets=df.groupby(["bowler","dismissal_kind"])["player_dismissed"].count()

print("Total wickets and their dismissal types by YS Chahal : ",wickets["YS Chahal"])
```

```python
#3) Top 3 runs scorers (all matches)
total=df.groupby("batsman")["total_runs"].sum()
#print(total)
total=total.sort_values(ascending=False) #descanding order

print("Top 3 runs scorers (all matches) : ",total.head(3))
```

```python
#4) The highest total score of an inning 
total_score_inning=df.groupby(['match_id', 'inning'])["total_runs"].sum()
print("The highest total score of an inning ",total_score_inning.max())
```
