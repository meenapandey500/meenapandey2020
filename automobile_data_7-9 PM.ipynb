```python
#1. MinMaxscaler method : how it is work : -
'''formula :  
X_new = (X_old-min(X))/(Max(X)-Min(X))'''
```

```python
#for example : -
#suppose we creates new 1D array a and b with the help of numpy array
a=np.array([10,20,30,40,50])
b=np.array([100,200,300,400,500])
```

```python
print(b)
```

```python
#create a function
def minmax(arr):  #passing argument /parameter and return value arr=[10,20,30,40,50]
    arr_new=(arr-arr.min())/(arr.max()-arr.min())
    return arr_new
minmax(a) #call function
```

```python
#Find outlier : There are found outlier from 2 methods : -
#1. IQR : Inter Quantile Range
#2. Z-score 
#zscore=(X-Mean(X))/std   where std means standard deviation
```

```python
#suppose We have data :  1,2,2,2,3,1,1,15,2,2,2,3,1,1,2 
#find outlier with the help of z-score
import numpy as np
```

```python
#create list data
data=[1,2,2,2,3,1,1,15,2,2,2,3,1,1,2]
#find mean of data
m=np.mean(data)
#find the standard deviation use inbuilt function std()
s=np.std(data)
print("Mean of our dataset : ",m)
print("Standard deviation of out dataset : ",s)
```

```python
#Calculate the Z-score . If Z-score>3 ,then print it as an outlier
threshold=3
outlier=[] #create empty list outlier
for i in data: 
    z=(i-m)/s
    z=abs(z) #abs() means absolute value
    print("z-score : ",z)
    if(z>threshold):
        outlier.append(i) #append() inbuilt function , to add value at list
        
print("Outlier found in dataset : ",outlier)
```
