```python
#conditional filtering : To access record according to condition from dataset
#Get all the rows where the bodystyle =hatchback 
df.columns
df[df["body-style"]=="hatchback"] #df[index] here conditions use for index
```

```python
#Get all the rows where the bodystyle =hatchback  and price>20000
#&  is used for & means apply multiple condition 

df[(df["price"]>20000) & (df["body-style"]=='hatchback')] #df[index] here conditions use for index
```

```python
#Get all the rows where the bodystyle =hatchback  or fuel-type=diesel
#Then use or operator ==> | pipeline
df[(df["fuel-type"]=='diesel') | (df["body-style"]=='hatchback')]
```

```python
#groupby
df.groupby("fuel-type").groups
```

```python
df.iloc[63] #access all details of record no. 63
```

```python
#group by creates groups of similar category and perform aggregate function
#Aggregate function means sum(),mean(),max(),min(),median()
```

```python
#What is the sum means Total price for each category of fuel type
df.groupby("fuel-type")["price"].sum()
```

```python
#What is the average means mean price for each category of fuel type
df.groupby("fuel-type")["price"].mean()
```

```python
df.groupby(["body-style","fuel-type"])["price"].sum() #list[] group([col1,col2,..])
```

```python
df.head()
```

```python
#What is average highway_milage for each manufacturer of cars.
df.groupby("make")["highway-mpg"].mean()
```

```python
#What is maximum highway_milage for each manufacturer of cars.
df.groupby("make")["highway-mpg"].max()
```

```python
#First create student dataframe
student=pd.DataFrame({'Rno':[110,120,130,140,150],'Name':["Anu","Seema","Rekha","Renu","Suresh"],
                     'Age':[21,23,24,25,20]})
```

```python
print(student)
```

```python
#Second Dataframe marks
marks=pd.DataFrame({'Rno':[110,130,140],'Phy':[67,78,89],'Maths':[89,90,65],'Eng':[56,76,78]})
```

```python
print(marks)
```

```python
X=pd.merge(student,marks,on="Rno") #Equal join/inner join bydefault how="equal"
#merge() ==>inbuilt function of pandas
#syntax : merge(dataframe1,dataframe2,on="common field")
```

```python
print(X)
```

```python
#outer join
X=pd.merge(student,marks,on="Rno",how="outer")
```

```python
print(X)
```

```python
#concat() ==> to concatenate means to join 2 series or dataframe
#concat in Series() means one dimensional array
s1=pd.Series(['a','b'])
s2=pd.Series(['c','d'])
print(s1)
print(s2)
```

```python
pd.concat([s1,s2])
```

```python
pd.concat([s1,s2],ignore_index=True)
```

```python
pd.concat([s1,s2],keys=['s1','s2'])
```

```python
pd.concat([s1,s2],keys=['s1','s2'],names=['Seires Name','Row Id'])
```
