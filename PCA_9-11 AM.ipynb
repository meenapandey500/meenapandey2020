```python
#Separate X and Y
X=df.drop("Customer_Segment",axis=1)
Y=df["Customer_Segment"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,
                                                     random_state=1)
```

```python
X_train.head()
```

```python
#scaling standardization : - on X_train and X_test
#inbuilt class StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
#fit_transform for col in df:
#create object of StandardScaler class
ss=StandardScaler()
X_train=ss.fit_transform(X_train) #apply on training data
X_test=ss.transform(X_test) #apply on testing data
```

```python
for col in X_train:
    print(col)
   
```

```python
for col in X_test:
    print(col)
```

```python
#PCA  : feature Extraction
```

```python
X_train.shape
```

```python
from sklearn.decomposition import PCA
```

```python
#Create object of PCA class
pc=PCA(n_components=None,random_state=1)  
#n_components  inbuilt parameters , here pass no. of pca but now given=None
#no pass any pca
```

```python
X_train_pc=pc.fit_transform(X_train) #Training data 
#fit_transform() inbuilt method of PCA
X_test_pc=pc.transform(X_test) #TEsting data
#transform() inbuilt method of PCA
```

```python
#to show strength means variance of each input column
explain_variance=pc.explained_variance_ratio_
print(explain_variance)
#explained_variance_ratio_ inbuilt method of pca class
```

```python
X.columns
```

```python
#Create object of PCA class
pc1=PCA(n_components=2,random_state=1)  
```

```python
X_train_pc1=pc1.fit_transform(X_train) #Training data 
#fit_transform() inbuilt method of PCA
X_test_pc1=pc1.transform(X_test) #TEsting data
#transform() inbuilt method of PCA
```

```python
#apply classification algorithm because we predict customer_segment
#means customer belong which class either 1 or 2 or 3
#then use classification algo.  : logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
#Create object of LogisticRegression
lr=LogisticRegression() #lr user defined object of LogisticRegression
```

```python
#Train the model then inbuilt function fit() of LogisticRegression class
lr.fit(X_train_pc1,Y_train)   #same as linearregression
```

```python
#find prediction means test the model
y_pred=lr.predict(X_test_pc1)
```

```python
#To generate report means show all scores
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,y_pred))
```

```python
#confusion matrix 
from sklearn.metrics import confusion_matrix
```

```python
#Confusion matrix
confusion_matrix(Y_test,y_pred)
```
