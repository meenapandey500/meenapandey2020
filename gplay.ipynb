```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sk ={'Class Marks':[61,64,67,70,73],'Frequency':[5,18,42,27,8]}
```

```python
df=pd.DataFrame(sk)
print(df)
```

```python
df.columns
```

```python
from scipy.stats import skew
```

```python
for col in df:
    print(col)
    print(skew(df[col]))
```

```python
for col in df:
    print(col)
    print(skew(df[col]))
    plt.figure()
    sns.distplot(df[col]) #visualise 
    plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
