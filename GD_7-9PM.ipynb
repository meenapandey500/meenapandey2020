```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#we create a numpy array 
x=np.array([1,2,3,4,5])  #independent means input variable
y=np.array([5,7,9,11,13]) #target variable means dependent variable
```

```python
def gradient_descent(x,y):  #user defined function
    b_curr=0 #suppose  intercept
    m_curr=0 #suppose  slope means gradient
    n=len(x) #no. of datapoints means no. of records in dataset
    iterations=1000
    learning_rate=0.01
    #print("m{} ,b{} ".format(m_curr,b_curr))
    print("m=",m_curr," b=",b_curr)
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations): #start i=0 stop=100-1=99 step=+1
        y_pred=m_curr*x+b_curr  #y=mx+b
        plt.plot(x,y_pred,color='green') #line
        mse=(1/n)*sum([val**2 for val in(y-y_pred)])#List comprehension
        #minimize the error (mse) , so differentiate mse w.r.p to m and b
        m_differate=-(2/n)*sum(x*(y-y_pred))
        b_differate=-(2/n)*sum(y-y_pred)
        #new value of m and b
        m_curr=m_curr- learning_rate*m_differate
        b_curr=b_curr- learning_rate* b_differate
        print("m= ",m_curr," b= ",b_curr," iteration : ",i," MSE : ",mse)
    return(y_pred)
```

```python
#call function
y_pred=gradient_descent(x,y)
print("Predicted price : ",y_pred)
```
