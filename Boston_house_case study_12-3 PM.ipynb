```python
#perform the model with Linear regression algorithm (Multiple Linear Regression)
#means many input and one output
```

```python
# select all input features from dataset (no apply data analysis)
#select input and output from dataframe 
X=df.drop("MEDV",axis=1) #to hold all columns apart from medv it is input features
Y=df["MEDV"] #target means output variable which want to predict
```

```python
#Divide the dataset into 70% and 30% for training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#we train the model with the help of LinearRegression class with 70% data
#call LinearRegression inbuilt class
from sklearn.linear_model import LinearRegression 
```

```python
#create object of class LinearRegression 
lr=LinearRegression() #lr user defined object of LinearRegression  class
```

```python
#we train the model use fit() with 70%: inbuilt method of LinearRegression class
lr.fit(X_train,Y_train)
```

```python
#check coefficient(m or slope or gradient or tan) and intercept(b)
lr.coef_  #coef_  inbuilt method of linearregression class
```

```python
X.columns
```

```python
#Intercept
lr.intercept_  #intercept_ inbuilt method of linearregression class
```

```python
#we test the model with 30% data :
#use predict() inbuilt method of LinearRegression class
#predict()   : Y_pred=m1X1+m2X2+m3X3+......+b
Y_pred=lr.predict(X_test)
```

```python
Y_test
```

```python
Y_pred
```

```python
#mean squared error 
from sklearn.metrics import mean_squared_error
#mean_squared_error inbuilt class 
#MSE=sum[(original output-predicted output)^2]/N
#N no. of records
```

```python
mean_squared_error(Y_test,Y_pred)
```

```python
#root mean squared error
mse=mean_squared_error(Y_test,Y_pred)
rmse=np.sqrt(mse)
print("root mean squared error : ",rmse)
```

```python
#Evaluation of model with the help of r2_score() inbuilt function
#r2_score() inbuilt function are defined in r2_score inbuilt class
from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)  #range of r2_score : 0 to +1 
#if r2_score >0.5 then ok 
#0.73 means 73 %  do correct prediction but try aprox. above 95% 
```

```python
#means good but not better because model is overfit 
```

```python
#perform model after data analysis in dataset
#Select input and output 
X=df[["LSTAT","RM","PTRATIO","TAX","INDUS"]] #input always 2D
Y=df["MEDV"] #output
```

```python
#Divide the dataset into 70% and 30% for training and testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#train the model with the linear regression class
#first create the object of linear regression
lr=LinearRegression()
```

```python
#We train the model with 70% data use fit() method
lr.fit(X_train,Y_train)
```

```python
#to show coefficient 
lr.coef_
```

```python
#check intercept
lr.intercept_
```

```python
#we test the model with 30% data :
#use predict() inbuilt method of LinearRegression class
Y_pred=lr.predict(X_test) #Y_pred=mX+b
```

```python
#root mean squared error
mse=mean_squared_error(Y_test,Y_pred)
print("Mean Squared Error : ",mse)
rmse=np.sqrt(mse)
print("root mean squared error : ",rmse)
```

```python
r2_score(Y_test,Y_pred)
```

```python
#Find residual
residuals=Y_test-Y_pred
```

```python
#we plot graph between residuals and y_pred
plt.figure()
sns.scatterplot(Y_pred,residuals)
plt.show()
```

```python
plt.figure()
sns.distplot(residuals)
plt.show()
```

```python
#It is just ok r2-score
```

```python
#assumption of Linear regression : -
#1. There should be no multicolinearity .
#2. There should be linear relationship between input and output
#here in this dataset Rm is good correlated with output MEDV
#so Rm and LSTAT both input are correlated each other bot both are good 
#correlation with output
```

```python
#polynomial Regression : -  means increase the degree of X 
#Y=mx+b   here increase the degree of X  Y=m1x^2+m2X+b 
# ax^2+bx+c=0
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Create the object of  PolynomialFeatures
pf= PolynomialFeatures(2) #pass How much degree 
```

```python
#We create  polynomial version of x so we call pf.fit_transform()
x_poly=pf.fit_transform(X)  #hold x and power of x means x^2
#whats return polynomial version of that means the degree 2 means find degree of x
```

```python
#train and test 
x_train,x_test,y_train,y_test=train_test_split(x_poly,Y,random_state=1,test_size=0.3)
```

```python
#Create the object of LinearRegression 
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#We can perform the prediction then call predict()
y_pred=lr.predict(x_test) 
```

```python
#Check mean squared error and r2-score
mean_squared_error(y_test,y_pred)   
```

```python
r2_score(y_test,y_pred)  #so great it is very close to 1 ,our assumption is correct 
#there is a polynomial relationship and because of great r2-score
```

```python
#calculate the residual 
residuals=y_test-y_pred
```

```python
#After calculation the residual,plot the residuals and predicted value
plt.figure()
sns.scatterplot(y_pred,residuals)
plt.show()
```

```python
plt.figure()
sns.distplot(residuals)
plt.show()
```

```python
#very close to Normal distribution 
#Finally concluted the great result so we can see that all assumption .
# There is a polynomial regression. here r2_score ==>is very good.
```
