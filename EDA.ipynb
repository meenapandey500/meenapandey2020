```python
#minmaxscaler method
```

```python
#x_new=(x_odd-min)/(max-min) ==>formula
```

```python
a=np.array([10,20,30,40,50]) #min of a=10 and max of a=50
```

```python
b=np.array([100,200,300,400,500]) #min of b=100 and max of b=500
```

```python
def minmax(ar): #user defined function
    ar_new=(ar-ar.min())/(ar.max()-ar.min()) #ar_new=(10-10)/40
    return ar_new
#main program
minmax(a)
minmax(b)
```
