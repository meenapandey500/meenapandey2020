```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load DataSet
df=pd.read_csv("train.csv")  #df user defined object
df.head() #to show top 5 records 
```

```python
df.isnull().sum()
```

```python
#plot 
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
plt.show()
```

```python
#To find the percentage of each column of dataset 
#the apply formula total null values of each column of dataset/no. of rows *100
#We wants to only row the df.shape[0]
null_var=df.isnull().sum()/df.shape[0]*100 
null_var
```

```python
#To check the missing values of those columns whose >17%
drop_columns=null_var[null_var>17].keys()
drop_columns
```

```python
#To remove those columns whose percentage of missing value>17%
df2_drop_cln=df.drop(columns=drop_columns) #drop() inbuilt function to remove column
df2_drop_cln.shape
```

```python
#Here 75 colums ,6 columns will be remove
```

```python
sns.heatmap(df2_drop_cln.isnull())
plt.show()
```

```python
#dropna() by default setting ==>row -->to remove those rows whose value is null
df3_drop_rows=df2_drop_cln.dropna()
```

```python
df3_drop_rows.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_drop_rows.isnull())
plt.show()
```
