# Apply Regularization

```python
from sklearn.linear_model import Lasso , Ridge
```

```python
#Lasso
#To check Lambda Value for Lasso 
for i in range(200,501,50):
    l1=Lasso(i)
    l1.fit(X_train,Y_train)
    print(i)
    print('Train-')
    print(l1.score(X_train,Y_train))
    print('Test-')
    print(l1.score(X_test,Y_test))
    print()
```

```python
#Ridge
#To check Lambda Value for Ridge 
for i in range(0,51,1):
    l2=Ridge(i)
    l2.fit(X_train,Y_train)
    print(i)
    print('Train-')
    print(l2.score(X_train,Y_train))
    print('Test-')
    print(l2.score(X_test,Y_test))
    print()
```

```python
# We will take lambda value 200 for Lasso and 2 for Ridge
```

```python
# Lasso - Lambda=200
l1=Lasso(200)
l1.fit(X_train,Y_train)
print('Train-')
print(l1.score(X_train,Y_train))
print('Test-')
print(l1.score(X_test,Y_test))
```

```python
l1.coef_
```

```python
# Ridge - Lambda=2
l2=Ridge(2)
l2.fit(X_train,Y_train)
print('Train-')
print(l2.score(X_train,Y_train))
print('Test-')
print(l2.score(X_test,Y_test))
```

```python
l2.coef_
```

# To check Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
#Cross Validation with Lasso
l1=Lasso(200)
l1_cross=cross_val_score(l1,x,y,cv=4)
print(l1_cross)
print(l1_cross.mean())
```

```python
#Cross Validation with Ridge
l2=Ridge(2)
l2_cross=cross_val_score(l2,x,y,cv=4)
print(l2_cross)
print(l2_cross.mean())
```

```python
# cross validation of both Lasso and Ridge are approximately same 
```
