```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#we create a numpy array 
x=np.array([1,2,3,4,5])  #independent means input variable
y=np.array([5,7,9,11,13]) #target variable means dependent variable
```

```python
def gradient_descent(x,y):
    b_curr=0 #suppose 
    m_curr=0 #suppose
    n=len(x)
    learning_rate=0.02 #range : 0.01 to 0.05  
    iterations=450
    print("m{} ,b{} ".format(m_curr,b_curr))
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations): #start i=0 stop=10-1=9 step=+1
        y_pred=m_curr*x+b_curr  #y=mx+b
        plt.plot(x,y_pred,color='green') #line
        cost=(1/n)*sum([val**2 for val in(y-y_pred)])#List comprehension
        #cost means mse : is also known as cost function
        mdiff=-(2/n)*sum(x*(y-y_pred))
        bdiff=-(2/n)*sum(y-y_pred)
        m_curr=m_curr- learning_rate*mdiff 
        b_curr=b_curr- learning_rate*bdiff
        #print("m{} ,b{} ,iteration{},MSE{}  ".format(m_curr,b_curr,i,cost))
        print("m= ",m_curr," b= ",b_curr," iteration : ",i," MSE : ",cost)
    return(y_pred)
```

```python
#call function
y_pred=gradient_descent(x,y)
print("Predicted price : ",y_pred)
```
