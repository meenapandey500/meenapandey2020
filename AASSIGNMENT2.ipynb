```python
df=df.dropna()
```

```python
df
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df,x="AGE",y="MEDV")
plt.show()
```

```python
features = ['LSTAT', 'RM'] #independent variables X
target = df['MEDV'] #dependent variables Y

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='+')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
#Highly -ve correlated with LSTAT AND MEDV AND highly +ve correlated with
#RM and MEDV
```

```python
plt.figure(figsize=(20, 5))

features = ['CRIM', 'ZN','RAD'] #independent variables X
target = df['MEDV'] #dependent variables Y

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='+')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
#CRIM,ZN and RAD are not correlated with output MEDV
```

```python
df.corr()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
X=df[["LSTAT","RM"]]#independent variable whose corelation is good with dependent variable
Y=df["MEDV"] #dependent variable
```

```python
#call class for train_test_split()
from sklearn.model_selection import train_test_split
```

```python
#Train_test_split : separate dataset into train and test split 70% & 30 %
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
#call the class
from sklearn.linear_model import LinearRegression
```

```python
#create object of LinearRegression class
lr=LinearRegression() #here lr user defined object of LinearRegression class
```

```python
#Train the model : use fit() method
lr.fit(X_train,Y_train)
```

```python
#we can perform the prediction (Test)
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Mean squared error
mse=mean_squared_error(y_test,y_pred)
```

```python
#root mean_squared_error
rmse=np.sqrt(mse)
```

```python
print("mean_squared_error :", mse)
print("Root mean_squared_error :",rmse)
```

```python
#evaluation method of LinearRegression then use R2-score inbuilt method :
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred) 
```
