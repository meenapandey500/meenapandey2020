```python
df["Price"].unique()
```

```python
#Change $ in price column will be removed 
df["Price"]=df["Price"].map(lambda X : X.lstrip("$").rstrip())
```

```python
#check : either remove $ sign or not
df["Price"].unique()
```

```python
df["Price"].value_counts()
```

```python
#To change the value of Everyone with new value 0 in Price column permanently
df["Price"]=df["Price"].replace('Everyone','0') 
#syntax : - df['columnname']=df['columnname'].replace("oldvalue","newvalue")
```

```python
#check
df["Price"].unique()
```

```python
#to check null values 
df.isnull().sum()
```

```python
df.info()
```

```python
#To change the datatype of Price column from object to Float permanently
df["Price"]=df["Price"].astype("float")
```

```python
#check
df.info()
```

```python
df["Size"].unique()
```

```python
#Remove M(unit M means Mega byte) in Size column
#we decided Change unit of size of kb to Mb
#so remove M in Size column
df['Size']=df['Size'].map(lambda X:X.rstrip('M'))
```

```python
#check either remove M or not
df["Size"].unique()
```

```python
# First remove k of Size variable then change the unit of Size from k to M,
#first change float type then divided by 1024 (because 1 MB=1024 KB) so,
#apply round function because value receive after deciaml 1 digit then 
#change in string type because Size column is object type
df['Size']=df['Size'].map(lambda X: str(round((float(X.rstrip('k'))/1024),1)) if X[-1]=='k' else X)
```

```python
#check
df['Size'].unique()
```

```python
df['Size'].value_counts()
```

```python
#Replace Varies with device  with null values(np.nan) permanently in Size column
df['Size']=df['Size'].map(lambda X:np.nan if X.startswith("Varies") else X)
```

```python
df['Size'].unique()
```

```python
#check null values
df.isnull().sum()
```

```python
df["Size"].value_counts()
```

```python
#Replace 1,000+ with null values in Size column
df['Size']=df['Size'].replace('1,000+',np.nan)
```

```python
#check null values
df.isnull().sum()
```

```python
df.info()
```

```python
#To change the data type of Size column from object to float
df["Size"]=df["Size"].astype("float")
```

```python
#check 
df.info()
```
