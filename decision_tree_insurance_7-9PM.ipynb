```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#To load dataset insurance.csv
df=pd.read_csv("insurance.csv")
```

```python
#select X (input features)and Y(target) from dataset df
X=df.drop("insuranceclaim",axis=1)
Y=df["insuranceclaim"]
```

```python
#train_test_split()
from sklearn.model_selection import train_test_split
```

```python
#Split our dataset in train test of 70% & 30%
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.metrics import classification_report
```

```python
def create_model(model): #user defined function
    model.fit(X_train,Y_train) #model train
    y_pred=model.predict(X_test)
    print(classification_report(Y_test,y_pred))
    return model
    
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create the object of DecisionTreeClassifier
dt1=DecisionTreeClassifier()
```

```python
#call function
dt1=create_model(dt1)
```

```python
#Output is very good means great that is perfect model
'''Decision Tree can automatic select the best features(columns).
For feature selection purpose ,Decision Tree are always used.
'''
```
