```python
#Visualization 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#We create dataframe of house price
#we create a list of Area and Price of house
area=[600,800,1000,1200,1500,1800]
price=[1000000,1200000,1500000,1800000,2200000,2500000]
#Create dataframe from list 
df=pd.DataFrame(list(zip(area,price)), columns=['Area','Price'])
df
```

```python
#syntax of scatter plot : to show relation between input and output variable
#plt.scatter(X,Y) #where X=independent variable means input and Y : dependent
#means output variable
X=df["Area"]
Y=df["Price"]
plt.scatter(X,Y,c="red",s=70,marker="*") #c for color bydefault blue color
#s is size of data point which unit is pixel .pixel means smallest unit
#of graphics 
plt.xticks(X)
plt.yticks(Y)
plt.xlabel("Area in Sq. Ft.")
plt.ylabel("Price in Lacs")
plt.show() 
```

```python
#line graph and scatter plot: -
#use plot() inbuilt function of matplotlib.pyplot class 
X=df["Area"] #independent variable means input
Y=df["Price"]#dependent variable means output
plt.scatter(X,Y,c="red",s=200,marker="*")
plt.plot(X,Y,c="green",linewidth=3)
plt.xticks(X)
plt.yticks(Y)
plt.xlabel("Area in Sq. Ft.")
plt.ylabel("Price in Lacs")
plt.show()
```

```python
#outlier : -
#create a numpy array
import numpy as np
```

```python
A=np.array([9,9,10,10,10,11,12,36])
print(A)
```

```python
import pandas as pd
```

```python
#To change 1D numpy array into Series of pandas(1 D)
B=pd.Series(A)
```

```python
print(B)
```

```python
#Find outlier 
#Find outlier
Q1=B.quantile(0.25) #quantile() inbuilt function means 0.25 means 25%
Q3=B.quantile(0.75) 
IQR=Q3-Q1  #find inter quantile range
lower_whisker=Q1-1.5*IQR
upper_whisker=Q3+1.5*IQR
print("First Quantile Q1 : ",Q1)
print("Third Quantile Q3 : ",Q3)
print("inter quantile range : ",IQR)
print("lower_whisker : ",lower_whisker)
print("upper_whisker : ",upper_whisker)
```

```python
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
#plot boxplot
sns.boxplot(data=B,x=B)
plt.xticks(B)
plt.show()
```

```python
B[B>12] #to show those records whoce >12
```

```python
#Remove outlier (outlier in 7th record)
B.drop(7,inplace=True)#inplace=True : take permanent action in dataset
#drop() inbuilt function to delete record from Series dataset B
#7th record no.
```

```python
#plot boxplot after remove outlier
sns.boxplot(data=B,x=B)
plt.xticks(B)
plt.show()
```
